```python
import os
import json
from PIL import Image
from tqdm import tqdm



def yolo_to_coco(IMG_ROOT_DIR, LABEL_ROOT_DIR, output_dir):
	# Define categories
	categories = [{'id': 0, 'name': 'chevrolet_malibu_sedan_2012_2016'},
               {'id': 1, 'name': 'chevrolet_malibu_sedan_2017_2019'},
               {'id': 2, 'name': 'chevrolet_spark_hatchback_2016_2021'},
               {'id': 3, 'name': 'chevrolet_trailblazer_suv_2021_'},
               {'id': 4, 'name': 'chevrolet_trax_suv_2017_2019'},
               {'id': 5, 'name': 'genesis_g80_sedan_2016_2020'},
               {'id': 6, 'name': 'genesis_g80_sedan_2021_'},
               {'id': 7, 'name': 'genesis_gv80_suv_2020_'},
               {'id': 8, 'name': 'hyundai_avante_sedan_2011_2015'},
               {'id': 9, 'name': 'hyundai_avante_sedan_2020_'},
               {'id': 10, 'name': 'hyundai_grandeur_sedan_2011_2016'},
               {'id': 11, 'name': 'hyundai_grandstarex_van_2018_2020'},
               {'id': 12, 'name': 'hyundai_ioniq_hatchback_2016_2019'},
               {'id': 13, 'name': 'hyundai_sonata_sedan_2004_2009'},
               {'id': 14, 'name': 'hyundai_sonata_sedan_2010_2014'},
               {'id': 15, 'name': 'hyundai_sonata_sedan_2019_2020'},
               {'id': 16, 'name': 'kia_carnival_van_2015_2020'},
               {'id': 17, 'name': 'kia_carnival_van_2021_'},
               {'id': 18, 'name': 'kia_k5_sedan_2010_2015'},
               {'id': 19, 'name': 'kia_k5_sedan_2020_'},
               {'id': 20, 'name': 'kia_k7_sedan_2016_2020'},
               {'id': 21, 'name': 'kia_mohave_suv_2020_'},
               {'id': 22, 'name': 'kia_morning_hatchback_2004_2010'},
               {'id': 23, 'name': 'kia_morning_hatchback_2011_2016'},
               {'id': 24, 'name': 'kia_ray_hatchback_2012_2017'},
               {'id': 25, 'name': 'kia_sorrento_suv_2015_2019'},
               {'id': 26, 'name': 'kia_sorrento_suv_2020_'},
               {'id': 27, 'name': 'kia_soul_suv_2014_2018'},
               {'id': 28, 'name': 'kia_sportage_suv_2016_2020'},
               {'id': 29, 'name': 'kia_stonic_suv_2017_2019'},
               {'id': 30, 'name': 'renault_sm3_sedan_2015_2018'},
               {'id': 31, 'name': 'renault_xm3_suv_2020_'},
               {'id': 32, 'name': 'ssangyong_korando_suv_2019_2020'},
               {'id': 33, 'name': 'ssangyong_tivoli_suv_2016_2020'}]

	# Initialize data dict
	data = {'train': [], 'validation': []}

	# Loop over splits
	for split in ['train2017', 'val2017']:
		split_data = {'info': {}, 'licenses': [], 'images': [], 'annotations': [], 'categories': categories}

		# Get image and label files for current split
		image_files = sorted(os.listdir(IMG_ROOT_DIR+split))
		label_files = sorted(os.listdir(LABEL_ROOT_DIR+split))

		# Loop over images in current split
		cumulative_id = 0
		with tqdm(total=len(image_files), desc=f'Processing {split} images') as pbar:
			for i, filename in enumerate(image_files):
				image_path = os.path.join(IMG_ROOT_DIR+split, filename)
				im = Image.open(image_path)
				im_id = i + 1

				split_data['images'].append({
					'id': im_id,
					'file_name': filename,
					'width': im.size[0],
					'height': im.size[1]
				})

				# Get labels for current image
				label_path = os.path.join(LABEL_ROOT_DIR+split, os.path.splitext(filename)[0] + '.txt')
				with open(label_path, 'r') as file:
					lines = file.readlines()

				for line in lines:
					values = list(map(float, line.strip().split(' ')))
					class_id = int(values[0])
					bbox_x = int(round(values[1]))
					bbox_y = int(round(values[2]))
     
					x_max, y_max = int(round(max(values[3], values[5], values[7]))), int(round(max(values[4], values[6], values[8])))

					bbox_width = x_max-bbox_x
					bbox_height = y_max-bbox_y

					split_data['annotations'].append({
						'id': cumulative_id,
						'image_id': im_id,
						'category_id': class_id,
						'bbox': [bbox_x, bbox_y, bbox_width, bbox_height],
						'area': bbox_width * bbox_height,
						'iscrowd': 0
					})

					cumulative_id += 1

				pbar.update(1)

		if split == 'train2017':
			data["train"] = split_data
		else:
			data["validation"] = split_data

	# Save data to JSON files
	for split in ['train', 'validation']:
		filename = os.path.join(output_dir, f'{split}.json')
		with open(filename, 'w') as f:
			json.dump({'data': data[split]}, f)

	return data

IMG_ROOT_DIR = '/home/kwy00/ysy/coco_style/'
LABEL_ROOT_DIR = '/home/kwy00/ysy/coco_style_anno/'
OUTPUT_DIR = '/home/kwy00/ysy/coco_style/annotations/'

coco_data = yolo_to_coco(IMG_ROOT_DIR, LABEL_ROOT_DIR, OUTPUT_DIR)
```
